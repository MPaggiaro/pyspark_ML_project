1. Write a Pandas program to join the two given dataframes along rows and assign all data.

```python
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

print("Original DataFrames:")
print(student_data1)
print("-------------------------------------")
print(student_data2)

concat_by_column = pd.concat([student_data1, student_data2], axis=0)
print(concat_by_column)
```

2. Write a Pandas program to join the two given dataframes along columns and assign all data.

```python
concat_by_rows = pd.concat([student_data1, student_data2], axis = 1)
print(concat_by_rows)
```

3. Write a Pandas program to append rows to an existing DataFrame and display the combined data.

```python
s6 = pd.Series(['S6', 'Scarlette Fisher', 205], 
               index=['student_id', 'name', 'marks'])

student_data = pd.concat([student_data1, s6])
print(student_data)
# wrong!

student_data = pd.concat([student_data1, pd.DataFrame([s6])])
print(student_data)
# partially right - index is wrong.

student_data = pd.concat([student_data1, pd.DataFrame([s6])], ignore_index= True)
print(student_data)

```

5. Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.

```python
# data:
exam_data = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

student_data = pd.concat([student_data1, student_data2])

merged_data = pd.merge(student_data, exam_data, on = "student_id", how = 'left')
print(merged_data)
```

6. Write a Pandas program to join the two dataframes using the common column of both dataframes.

```python
# joining student data 1 and 2 on the same student ID:

conflicting_student_ids = pd.merge(student_data1, student_data2,
                                   on="student_id")
# if not specified, the merge is an inner join, i.e. keeps only entries
# that are present in both dfs.
print(conflicting_student_ids)
```

8. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})

merged_data = pd.merge(data1, data2, on = ['key1', 'key2'], how = 'left')
print(merged_data)
```
